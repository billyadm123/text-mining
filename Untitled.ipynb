```python
f = 'trump_georgia_transcripts.txt'
```

```python
idx = 0
word_dict = {}
# tokenized_docs = []
for line in open(f):
    tokens = line.split()
#     doc_to_int = []
    for i in tokens:
        if i not in word_dict:
            word_dict[i] = idx
            idx += 1
#         doc_to_int.append(word_dict[i])
#     tokenized_docs.append(doc_to_int)
```

```python
idx_word = {v:k for k, v in word_dict.items()}
```

```python
word_dict
```

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
nltk.download('stopwords')
stops = stopwords.words('english')
```

```python
stop_words = []
for i in word_dict:
    if i in stops:
        stop_words.append(i)
        
```

```python
stop_words
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

stops = set(STOPWORDS)

wordcloud = WordCloud(min_font_size=3, stopwords=stops).generate(str(word_dict.items()))

plt.figure(figsize=(30,29))
plt.imshow(wordcloud, interpolation='bilinear')
plt.title('With Stopwords', fontsize=40)
plt.axis("off")
plt.show()
```

```python
new_list= [i for i in word_dict if i not in stop_words]
```

```python
wordcloud = WordCloud(min_font_size=3, stopwords=stops).generate(str(new_list))

plt.figure(figsize=(30,29))
plt.imshow(wordcloud, interpolation='bilinear')
plt.title('Without Stopwords', fontsize=40)
plt.axis("off")
plt.show()
```
